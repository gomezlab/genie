```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('../data/klaeger/klaegerscan_data_for_CRC_patient_model.csv', index_col=0)
data.head()
```

```python
drugs = data.index.to_list()
ola_drugs = [drug for drug in drugs if 'olaparib' in drug]
ola_drugs
```

```python
act_cols = [col for col in data.columns if 'act_' in col]
len(act_cols)
```

```python
clin = pd.read_csv('../data/crc_ib_mut_cna_fus_clin.csv', index_col=0)
clin.head()
```

```python
for col in act_cols:
    clin[col] = np.NaN
```

```python
clin.head()
```

```python
print(clin.shape)
```

```python
#fill the act_ cols with the values from the klaeger data, matching on id_drugs_list in both dataframes
for i, row in clin.iterrows():
    for col in act_cols:
        if row['id_drugs_list'] in data['id_drugs_list'].values:
            clin.loc[i, col] = data.loc[data['id_drugs_list'] == row['id_drugs_list'], col].values[0]
        else:
            print('id_drugs_list not found in klaeger data: ', row['id_drugs_list'])
clin.head()
```

```python
#if everolimus or olaparib is in 'id_drugs_list', drop the row
print(clin.shape)
clin = clin[~clin['id_drugs_list'].str.contains('everolimus', case=False, na=False)]
clin = clin[~clin['id_drugs_list'].str.contains('olaparib', case=False, na=False)]
print(clin.shape)
```

```python
fus_cols = [col for col in clin.columns if 'fus_' in col]
len(fus_cols)
```

```python
clin.to_csv('../data/crc_ib_mut_cna_fus_act_clin.csv')
```
