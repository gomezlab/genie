```python
import pandas as pd
import numpy as np
```

```python
all_out = pd.read_csv('../data/crc_alltx_out.csv', index_col=0)
all_out.head()
```

```python
all_out.columns = ['id_' + i if i not in ['PFS', 'OS'] else i for i in all_out.columns]
```

```python
clin = pd.read_csv('../data/crc_clin_pub.csv', index_col=0)
clin.head()
```

```python
clin = clin.loc[clin['stage_dx'] == 'Stage II']
print(clin.shape)
clin.head()
```

```python
clin.columns = ['clin_' + i if i not in ['PFS', 'OS', 'OS_time'] else i for i in clin.columns]
cat_cols = [col for col in clin.columns if col.startswith('clin') and clin[col].nunique() <= 10]
num_cols = [col for col in clin.columns if col.startswith('clin') and clin[col].nunique() > 10]
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for i in cat_cols:
    clin[i] = le.fit_transform(clin[i])
from sklearn.impute import SimpleImputer
imp = SimpleImputer(strategy='median')
for i in num_cols:
    clin[i] = imp.fit_transform(clin[i].values.reshape(-1,1))
```

```python
mut = pd.read_csv('../data/crc_mutoh_pertreat.csv', index_col=0)
mut.columns = ['mut_' + i for i in mut.columns]
cna = pd.read_csv('../data/crc_cna_pertreat.csv', index_col=0)
cna.columns = ['cna_' + i for i in cna.columns]
fusion = pd.read_csv('../data/crc_fusion.csv', index_col=0)
fusion.columns = ['fus_' + i for i in fusion.columns]
```

```python
all_out.head()
```

```python
all_out = all_out.set_index('id_sample_id')
s2_mut = mut.join(all_out, how='inner')
s2_mut_cna = cna.join(s2_mut, how='inner')
s2_mut_cna_fus = fusion.join(s2_mut_cna, how='inner')
s2_mut_cna_fus.shape
```

```python
s2_mut_cna_fus.head()
```

```python
clin_pts = clin.index.tolist()
len(clin_pts)
```

```python
s2_mut_cna_fus = s2_mut_cna_fus.loc[s2_mut_cna_fus['id_record_id'].isin(clin_pts)]
s2_mut_cna_fus.shape
```
