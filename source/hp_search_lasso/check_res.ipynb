```python
import pandas as pd
```

```python
data_type = ['clin', 'nonclin', 'comb']
models = ['rf', 'xgb', 'nn']
res_dict = {}
for dt in data_type:
    for model in models:
        res_dict[dt + '_' + model] = pd.read_csv('../../results/lasso_runs/102123/results_' + model + '_folfox_OS_' + dt + '_102123.csv')
```

```python
res_dict['nonclin_rf'].head()
```

```python
res_dict['nonclin_nn'].head()
```

```python
#make a dataframe with all the results
res_df = pd.DataFrame(columns=['data_type', 'model', 'val_auroc', 'test_auroc'])
for dt in data_type:
    for model in models:
        #for val_auroc and test_auroc, take the mean of the 5 runs
        val_auroc = res_dict[dt + '_' + model]['val_auroc_mean'].mean()
        test_auroc = res_dict[dt + '_' + model]['test_auroc_mean'].mean()
        #append is deprecated, so use concat
        res_df = pd.concat([res_df, pd.DataFrame([[dt, model, val_auroc, test_auroc]], columns=['data_type', 'model', 'val_auroc', 'test_auroc'])], ignore_index=True)

res_df
```
