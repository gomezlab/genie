```python
import pandas as pd
```

```python
data_folfox = pd.read_csv('../data/crc_folfox_mut_cna_fus_clin.csv', index_col=0)
data_folfox.head()
```

```python
data_folfiri = pd.read_csv('../data/crc_folfiri_mut_cna_fus_clin.csv', index_col=0)
data_folfiri.head()
```

```python
#concatenate the two datasets
data = pd.concat([data_folfox, data_folfiri], axis=0)
print(data_folfox.shape, data_folfiri.shape, data.shape)
```

```python
#find the number of duplicate samples (based on the index)
print(data.index.duplicated().sum())
```

```python
#if the index is duplicated, and 'id_drugs_list' contains 'irinotecan' drop the sample
data = data[~(data.index.duplicated() & data['id_drugs_list'].str.contains('Irinotecan'))]
print(data.shape)
```

```python
#find the number of duplicate samples (based on the index)
print(data.index.duplicated().sum())
```

```python
#drop duplicates based on the index, keep the first sample
data = data[~data.index.duplicated(keep='first')]
print(data.shape)
```

```python
data.to_csv('../data/crc_folfox_folfiri_mut_cna_fus_clin.csv')
```
