```python
import pandas as pd
import numpy as np
```

```python
fusion = pd.read_csv('../data/crc_raw/cBioPortal_files/data_fusions.txt', sep='\t')
fusion.head()
```

```python
len(fusion['Hugo_Symbol'].unique())
```

```python
gene_list = fusion['Hugo_Symbol'].unique()
sample_list = fusion['Tumor_Sample_Barcode'].unique()
len(sample_list)
```

```python
data = pd.DataFrame(index=sample_list, columns=gene_list)
```

```python
for i, row in fusion.iterrows():
    data.loc[row['Tumor_Sample_Barcode'], row['Hugo_Symbol']] = 1
```

```python
data.head()
```

```python
data.fillna(0, inplace=True)
data.head()
```

```python
data.to_csv('../data/crc_fusion.csv')
```
